# Task 1 — Data Parsing & Profile Cleaning

```python
raw_students = [
    {"name": "  ayesha SHARMA  ", "roll": "101", "marks_str": "88, 72, 95, 60, 78"},
    {"name": "ROHIT verma",       "roll": "102", "marks_str": "55, 68, 49, 72, 61"},
    {"name": "  Priya Nair  ",    "roll": "103", "marks_str": "91, 85, 88, 94, 79"},
    {"name": "karan MEHTA",       "roll": "104", "marks_str": "40, 55, 38, 62, 50"},
    {"name": " Sneha pillai ",    "roll": "105", "marks_str": "75, 80, 70, 68, 85"},
]
```

```python
cleandata_students = []

for x in raw_students:
    # Name: Remove spaces and update to Title case
    name = x["name"].strip().title()
    
    # Roll: Converted to Int
    roll = int(x["roll"])

    # marks_str: Mark integers created
    mark = list(map(int, x["marks_str"].split(", ")))

    #Add the cleaned data to cleandata_students
    cleandata_students.append({
        "name": name,
        "roll": roll,
        "marks_list": mark
    })
```

```python
#Loop through Cleanedata_students name field and store the validation in 'nameValid' bool
for x in cleandata_students:
    for n in x["name"].split():
        if n.isalpha():
            nameValid = True
    #Print Name: Validity message
    if nameValid:
        print (x["name"] + ": ✓ Valid name")
    else:
        print (x["name"] + ": ✗ Invalid name")
```

```python
#Profile Card for each cleaned student
for x in cleandata_students:
    print ("================================")
    print("Student : " + x["name"])
    print(f"Roll No : { x["roll"]}")
    print(f"Marks   : { x["marks_list"]}")
```

```python
# Find the student with roll number 103 and print in required format
for student in cleandata_students:
    if student["roll"] == 103:
        print (student["name"].upper())
        print (student["name"].lower())
```

# Task 2 — Marks Analysis Using Loops & Conditionals

```python
student_name = "Ayesha Sharma"
subjects     = ["Math", "Physics", "CS", "English", "Chemistry"]
marks        = [88, 72, 95, 60, 78]
```

```python
# Assign grading based on marks and return them for analysis
def checkGrade(mark):
    if 90 <= mark <=100:
        return "A+"
    elif 80 <= mark <= 89:
        return "A"
    elif 70 <= mark <= 79:
        return "B"
    elif 60 <= mark <= 69:
        return "C"
    else:
        return "F"


for s,m in zip(subjects,marks):   
    grade = checkGrade(m)
    print(f"{s:<10} : {m} : {grade}")
```

```python
# Grade Calculations

# Print Total marks
print(f" Total Marks : {sum(marks)}")

# Print Avergae marks
print(f" Average Marks : {round(sum(marks) / len(marks), 2)}")

# Highest Scoring Subject
print(f" Highest scoring subject : {subjects[marks.index(max(marks))]}  {max(marks)}")

# Lowest Scoring Subject
print(f" Lowest scoring subject :  {subjects[marks.index(min(marks))]}  {min(marks)}")
    
```

## Marks-entry system that allows addition of new subjects:

```python
newSub = 0

print("Add new subjects (type 'done' to stop):")

while True:
    sub = input("Enter subject name: ")
    
    if sub.casefold() == "done":
        break
    
    marks_new = input("Enter marks (0-100): ")
    
    # Validate if marks is digit
    if not marks_new.isdigit():
        print("Invalid input! Marks must be a number.")
        continue
    
    mark = int(marks_new)

    # Validate if marks is outside valid range
    if mark < 0 or mark > 100:
        print("Invalid input! Marks must be between 0 and 100.")
        continue
    
    # Add valid data
    subjects.append(sub)
    marks.append(mark)
    newSub += 1

# Final updated average
updated_avg = round(sum(marks) / len(marks), 2)

print(f"New Subjects Added : {newSub}")
print(f"Updated Average    : {updated_avg}")
```

# Task 3 — Class Performance Summary

```python
class_data = [
    ("Ayesha Sharma",  [88, 72, 95, 60, 78]),
    ("Rohit Verma",    [55, 68, 49, 72, 61]),
    ("Priya Nair",     [91, 85, 88, 94, 79]),
    ("Karan Mehta",    [40, 55, 38, 62, 50]),
    ("Sneha Pillai",   [75, 80, 70, 68, 85]),
]
```

```python
# Average of marks for each student
print( "Name              | Average | Status")

# Calculate the number of passed and failed students for later analysis
pass_count =0
fail_count =0
averageMarks = []

# To find out the topper name and average
avgTopper = 0

for name,mark in class_data:
    average = round(sum(mark)/len(mark),2)
    averageMarks.append(average) #Store all averages in a list
    if average >=60:
        result = "Pass"
        pass_count += 1
    else:
        result = "Fail"
        fail_count += 1
    if average > avgTopper:
        avgTopper = average
        nameTopper = name

    print( "----------------------------------------")
    print(f"{name:<18}| {average:^7} | {result}")
```

```python
# Result analysis:

print(f"Number of students passed: {pass_count}")
print(f"Number of students failed: {fail_count}")
print(f"Class Topper: {nameTopper}, Average: {avgTopper}")
print(f"Class Average: {round(sum(averageMarks)/len(averageMarks),2)}")
```

```python
essay = "  python is a versatile language. it supports object oriented, functional, and procedural programming. python is widely used in data science and machine learning.  "
```

```python
# Clean the essay 

clean_essay = essay.strip()
```

```python
# Convert cleaned essay to display Title Case
titlecase_essay = clean_essay.title()

print("Title Case: ")
print(titlecase_essay)
```

```python
# Calculating how many times the word 'python' appears

count = clean_essay.count("python")
print(f"Count of word 'python': {count}")
```

```python
replaced_essay = clean_essay.replace("python", "Python 🐍")
print("Replaced Essay:")
print(replaced_essay)
```

```python
split_essay = clean_essay.split(". ")
print("Sentences List:")
print(split_essay)
```

```python
print("Numbered Sentences:")
for i, sentence in enumerate(split_essay, start=1):
    # Ensure sentence ends with "."
    if not sentence.endswith("."):
        sentence += "."
    print(f"{i}. {sentence}")
```
